Below is just to create a test traiing data, and plot a graph

```python
import numpy as np
import matplotlib.pyplot as plt



x_train = np.array([1000,1100,1200])
y_train = np.array([100,110,120])


plt.scatter(x_train, y_train,marker = 'x', c ='r')
plt.title("Housing Prices Chart")
plt.ylabel("Prices in K Dollars")
plt.xlabel("Area in sqft")
plt.show()
```

Creating a training model with input params as input variable (x_train) , weight(w) and bias (b) , using liner regression

```python
def compute_model(x,w,b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = (w * x[i]) + b
  return f_wb
```

Training the model using compute_model function

```python
w = 10
b = 10
y_val = compute_model(x_train,w,b)

print(y_val)
```

Plottig a a linear graph using the training model with random weight (w) and bias (b)

```python
plt.plot(x_train,y_val,c = 'b', label = 'Our Prediction')

plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title("Housing Prices Chart")
plt.ylabel("Prices in K Dollars")
plt.xlabel("Area in sqft")
```
