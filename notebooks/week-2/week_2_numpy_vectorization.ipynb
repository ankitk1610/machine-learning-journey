Slicing

```python
array = np.arange(10)

print(array[-1])
# Slice from 1 till 8 (1 inclusive, 8 exclusive, with step size 2)
print(array[1:8:2])

# Slice from start till 6 ,  with step size 4
print(array[:6:4])

# Slice from 6 till end ,  with step size 4
print(array[6::4])
```

Demo for dot product with and without vectorise dot product using numpy

Definfing dot product iteratively

```python
def dot_product(a,b):
  ans = 0;
  for i in range(a.shape[0]):
    ans = ans + (a[i]*b[i])
  return ans
```

```python
a = np.arange(10)
b = np.arange(10)

print(dot_product(a,b))

print(np.dot(a,b))
```

Comparing time

```python
np.random.seed(1)


a = np.random.rand(100000)
b = np.random.rand(100000)


tic = time.time()
d = dot_product(a,b)
toc = time.time()

print("Using dot product without numpy")
print(d)
print(toc-tic)


print("Using dot product with numpy")

tic = time.time()
d = np.dot(a,b)
toc = time.time()

print(d)
print(toc-tic)

```

```python
sizes = np.array([1,10,100,1000,10000,100000,1000000,10000000,100000000])

without_vectorise = np.zeros(sizes.shape[0])
with_vectorise = np.zeros(sizes.shape[0])
for i in range(sizes.shape[0]):
  a = np.random.rand(sizes[i])
  b = np.random.rand(sizes[i])
  tic = time.time()
  d = dot_product(a,b)
  toc = time.time()
  without_vectorise[i] = toc-tic

  tic = time.time()
  d = np.dot(a,b)
  toc = time.time()
  with_vectorise[i] = toc-tic



print(without_vectorise)
print(with_vectorise)





```

Plotting Graphs for comparison

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(sizes, without_vectorise, label="Without Vectorization")
plt.plot(sizes, with_vectorise, label="With Vectorization")

plt.xlabel("Input Size")
plt.ylabel("Time (seconds)")
plt.title("Dot Product Performance Comparison")
plt.legend()

# Optional but useful for large input variation
plt.xscale("log")
plt.yscale("log")

plt.grid(True)
plt.show()
```
